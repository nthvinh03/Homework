```python
cleaned_data=df[df['date']=='2020-04-12']
```

```python
plt.hist(cleaned_data.cases,bins=200)
```

```python
print("Tổng số ca nhiễm và số ca tử vong ở mỗi châu Lục")
cleaned_data.groupby('continent')[['cases','deaths']].sum()
```

```python
print("5 quốc gia có số ca mắc nhiều nhất")
cleaned_data=cleaned_data.sort_values('cases',ascending=False)
cleaned_data.head()
```

```python
print("Top 5 quốc gia có số ca tử vong cao nhất")
cleaned_data=cleaned_data.sort_values('cases',ascending=False)
cleaned_data.head()
```
