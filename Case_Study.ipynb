# Tìm hiểu mối liên hệ giữa hãng xe và tên xe 

```python
df['Car Brand']=df['CarName'].str.split(' ',expand=True)[0]
df['Car Name']=df['CarName'].str.split(' ',expand=True)[1]
```

# Xem xét kiểu dữ các thuộc tính và chuyển đổi về đúng kiểu 

```python
df['doornumber'].unique()
```

```python
door_number={
    'two':2,
    'four':4
}
```

```python
df['doornumber']=df['doornumber'].replace(door_number)
df['doornumber'].astype(int)
```

```python
df['cylindernumber'].unique()
```

```python
cylinder_number={
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'eight':8,
    'twelve':12
}
```

```python
df['cylindernumber']=df['cylindernumber'].replace(cylinder_number)
df['cylindernumber'].astype(int)
```

```python
df[['doornumber','cylindernumber']].head()
```

# Phân tích mối liên hệ của từng thuộc tính với Price

```python
#price vs fueltype

fueltype_price=df.groupby('fueltype')['price'].mean().reset_index()

plt.bar(fueltype_price['fueltype'],fueltype_price['price'])

# Giá xe sử dụng động cơ diesel cao hơn giá xe sử dụng gas
```

```python
brand_price=df.groupby('Car Brand')['price'].mean().reset_index()
sns.barplot(brand_price,x='Car Brand',y='price')
```

```python
fix,axs=plt.subplots(1,8,figsize=(20,5))
sns.regplot(x='symboling',y='price',data=df,ax=axs[0])
sns.regplot(x='doornumber',y='price',data=df,ax=axs[1])
sns.regplot(x='carlength',y='price',data=df,ax=axs[2])
sns.regplot(x='cylindernumber',y='price',data=df,ax=axs[3])
sns.regplot(x='wheelbase',y='price',data=df,ax=axs[4])
sns.regplot(x='horsepower',y='price',data=df,ax=axs[5])
sns.regplot(x='carwidth',y='price',data=df,ax=axs[6])
sns.regplot(x='carheight',y='price',data=df,ax=axs[7])
plt.show()
```
